```python
Lshow('If GDC between:')
P=MyEq((x*x-2*x+1)**4*(x*x++3*x+2)**8,'P')
Q =MyEq((x*x*x-3*x*x+3*x-1)**4*(x+2)**8,'Q')
Lshow(' Is:')
M=MyEq((a*x*x+b*x+c)**n,'M')
Lshow(' find : abc+n')
```

```python
P.factor()
```

```python
Q.factor()
```

```python
GCD(P,Q) # GDC(a,b) return gdc between (a,b)
```

```python
# GDC(a,b) return gdc between (a,b)
# P.baselist() : if P= (x+a)**a*(y-2)**b*z**c,  return [(x+a),(y-2),z]
# mulitem(ksym): input[a,b,c...] return a*b*c*...
# P=MyEq(expr,'P') return Equation class with name P
A=MyEq(mulitem(baselist(GCD(P,Q))),'A')
```

```python
# Q=MQ(expr1,expr2) return Equation Equallity  class with name Q--> expr1=expr2
B=MQ(A,getbase(M))
```

```python
B.expand()
```

```python
a,b,c=B.solve_coef_list(a,b,c)
```

```python
a*b*c+8
```
